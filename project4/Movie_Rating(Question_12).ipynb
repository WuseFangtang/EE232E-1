```python
import numpy as np
import pandas as pd
import statsmodels.formula.api as smf
from sklearn.metrics import mean_squared_error
import statsmodels.tools.eval_measures as smm
```

```python
movie_rating = pd.read_table("movie_rating.txt",encoding="ISO-8859-1",names=["Movie","NA","Rate"]).drop("NA",axis=1)
```

```python
movie_rating.to_csv("movie_rating(new).txt")
```

```python
movie_rate_dict = movie_rating.to_dict('list')
movies = movie_rate_dict['Movie']
rates = movie_rate_dict['Rate']
movie_rate_dict = dict(zip(movies,rates))
```

```python
actor_score = pd.read_table("actor_score.csv",float_precision='high',sep=",",encoding="ISO-8859-1",skiprows=1,names=["No.","Actor","Score"]).drop("No.",axis=1)
```

```python
actor_score = actor_score.set_index("Actor")
actor_score = actor_score.rename_axis(None)
```

```python
actor_score
```

```python
actor_score_dict = actor_score.to_dict()['Score']
```

```python
actor_score_dict['Cruise, Tom']
```

```python
import json
f = open('actor_movie_clean.txt', 'r')
movie_actors_dict = json.loads(f.read())
f = open('director_movies_clean.txt', 'r')
director_movies_dict = json.loads(f.read())
f = open('movie_directors_clean.txt', 'r')
movie_directors_dict = json.loads(f.read())
f.close()
```

```python
movie_actors_dict['Batman v Superman: Dawn of Justice (2016)']
```

```python
def Avg_Pagerank(MovieName):
    if MovieName not in movie_actors_dict:
        return None
    
    return np.mean([actor_score_dict[actor] for actor in movie_actors_dict[MovieName] if actor in actor_score_dict]) 
```

```python
movie_rating['Actors_Pagerank_Avg']= movie_rating['Movie'].apply(Avg_Pagerank)
```

```python
director_movies_dict
```

```python
director_score_dict = {}
for director,movies in director_movies_dict.items():
    director_score_dict[director] = np.mean([movie_rate_dict[movie] for movie in movies if movie in movie_rate_dict])
        
director_score_dict
```

```python
movie_directors_dict
```

```python
def Max_Director_Score(MovieName):
    if MovieName not in movie_directors_dict:
        return None
    
    Directors = movie_directors_dict[MovieName]
    result = np.max([director_score_dict[director] for director in Directors if director in director_score_dict])
    
    return result

def Category_Director_Score(MovieName):
    if MovieName not in movie_directors_dict:
        return None
    
    Directors = movie_directors_dict[MovieName]
    result = np.max([director_score_dict[director] for director in Directors if director in director_score_dict])
    
    if result > 8:
        return 8
    elif result >7.5:
        return 7
    elif result >7:
        return 6
    elif result >6.5:
        return 5
    elif result >6:
        return 4
    elif result >5.5:
        return 3
    elif result >5:
        return 2
    else:
        return 1
```

```python
movie_directors_dict['Brigády (1947)']
movie_rating['Max_Director_Score']= movie_rating['Movie'].apply(Max_Director_Score)
```

```python
movie_rating
```

```python
movie_rating_cleared=movie_rating.dropna()
movie_rating_cleared
```

```python
#X = movie_rating_cleared[['Max_Director_Score']]
X = movie_rating_cleared[['Actors_Pagerank_Avg','Max_Director_Score']]
Y = movie_rating_cleared['Rate']
```

```python
model = smf.OLS(Y, X).fit()
predictions = model.predict(X)
```

```python
smm.rmse(Y, predictions)
```
