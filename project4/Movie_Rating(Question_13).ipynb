```python
import json
import numpy as np
import pandas as pd
from igraph import Graph
from heapq import nlargest
import statsmodels.tools.eval_measures as smm
```

```python
f = open('actor_movie_clean.txt', 'r')
movie_actors_dict = json.loads(f.read())
f.close()
```

```python
movie_actors_dict
```

```python
edges = []
actors_set = set()
for movie,actors in movie_actors_dict.items():
    for actor in actors:
        edges.append((actor,movie))
        actors_set.add(actor)
```

```python
actors_list = list(actors_set)
actors_list
```

```python
bigraph = Graph.TupleList(edges, directed = False)
bigraph
```

```python
bigraph.vs.find("Cube, Ice").index
```

```python
movie_rating = pd.read_table("movie_rating.txt",encoding="ISO-8859-1",names=["Movie","NA","Rate"]).drop("NA",axis=1)
movie_rate_dict = movie_rating.to_dict('list')
movies = movie_rate_dict['Movie']
rates = movie_rate_dict['Rate']
movie_rate_dict = dict(zip(movies,rates))
movie_rate_dict
```

```python
for actor in actors_list:
    actor_node = bigraph.vs.find(actor)
    movies_involved_id = bigraph.neighbors(actor_node) #ID numbers of movie_node
    movies_involved = [bigraph.vs[id]['name'] for id in movies_involved_id]
    movie_rate = [movie_rate_dict[movie] for movie in movies_involved if movie in movie_rate_dict]
    
    top_10 = nlargest(10,movie_rate)
    if len(movie_rate)>0:
        actor_node['Value'] = np.mean(movie_rate)   
    else:
        actor_node['Value'] = None
```

```python
predict_list=['Batman v Superman: Dawn of Justice (2016)',
                     'Mission: Impossible - Rogue Nation (2015)',
                     'Minions (2015)']
```

```python
for movie in predict_list:
    movie_node = bigraph.vs.find(movie)
    actors_involved_id = bigraph.neighbors(movie_node) #ID numbers of actor_node
    movie_rate = [bigraph.vs[actorid]['Value'] for actorid in actors_involved_id if bigraph.vs[actorid]['Value']!=None]
    print("The predict rate for {} is {}".format(movie,np.mean(movie_rate)))
```

```python
predict_rates = {}
for movie in movie_rating['Movie']:
    if movie in movie_actors_dict:
        actors_involved_id = bigraph.neighbors(movie) #ID numbers of actor_node
        movie_rate = [bigraph.vs[actorid]['Value'] for actorid in actors_involved_id if bigraph.vs[actorid]['Value']!=None]
        #print("The predict rate for {} is {}".format(movie,np.mean(movie_rate)))
        top_10 = nlargest(10,movie_rate)
        predict_rates[movie] = np.mean(movie_rate)
```

```python
keys_predict = set(predict_rates.keys())
keys = set(movie_rate_dict.keys())
intersection = keys_predict & keys
```

```python
smm.rmse([movie_rate_dict[i] for i in intersection], [predict_rates[i] for i in intersection])
```
